**3.** Write a program that calculates the factorial of a given number. The code below is only an example programmed in a recursive way that can be used to compare with the one you get.  

```python
N = 5
#Factorial function
def factorial(N):
    if N==1:
        return 1
    else:
        return N*factorial(N-1)
    
print "The factorial of %d is %d"%(N, factorial(N))
```
